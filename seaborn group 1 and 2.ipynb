```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style = "whitegrid")
```

```python
Salary_data = pd.read_csv("Salary Data.csv")
```

```python
Salary_data.head()
```

```python
plt.figure() 
sns.scatterplot(x="Age", y= "Salary", data = Salary_data)
plt. title("Default Scatter Plot: Age vs Salary")
plt.show()
```

```python
plt.figure()
sns.scatterplot(x = "Age", y = "Salary", hue = "Education Level", style = "Gender", data = Salary_data)
plt.title("Scatter Plot: Age vs Salary by Education Level and Gender")
plt.show()
```

```python
education_colors = {"Bachelor's" : "green", "Master's" : "blue", "PhD" : "red"}
plt.figure()
sns.scatterplot(x = "Age", y = "Salary", hue = "Education Level", style = "Gender", data = Salary_data, palette=education_colors)
plt.title("Scatter Plot: Age vs Salary by Education Level and Gender")
plt.show()
```

```python
education_colors = {"Bachelor's" : "green", "Master's" : "blue", "PhD" : "red"}
gender_markers = {'Female':"o", "Male": "s"}

plt.figure()
sns.scatterplot(x = "Age", y = "Salary", hue = "Education Level", style = "Gender", data = Salary_data, palette=education_colors)
plt.title("Scatter Plot: Age vs Salary by Education Level and Gender")
plt.show()
```

```python
education_colors = {"Bachelor's" : "green", "Master's" : "blue", "PhD" : "red"}
gender_markers = {'Female':"o", "Male": "s"}

plt.figure(figsize = (12,8))
sns.set_style("ticks")
sns.scatterplot(x = "Age", y = "Salary", hue = "Education Level", style = "Gender", data = Salary_data, palette=education_colors)
plt.title("Scatter Plot: Age vs Salary by Education Level and Gender")
plt.show()
```

```python
education_colors = {"Bachelor's" : "green", "Master's" : "blue", "PhD" : "red"}
gender_markers = {'Female':"o", "Male": "s"}

plt.figure(figsize = (12,8))
sns.set_style("ticks")
sns.scatterplot(x = "Age", y = "Salary", hue = "Education Level", style = "Gender", data = Salary_data, palette=education_colors, markers = gender_markers)
plt.legend(ncol=2, frameon=False, loc= "lower right")
plt.title("Scatter Plot: Age vs Salary by Education Level and Gender")
plt.show()
```

```python

gender_colors = {'Female':"red", "Male": "blue"}

plt.figure(figsize = (12,8))
sns.set_style("ticks")
Bachelors = sns.scatterplot(x = "Age", y = "Salary", hue = "Gender", data = Salary_data[Salary_data["Education Level"] == "Bachelor's"], palette=gender_colors)
Bachelors.set_xlim(25,50)
Bachelors.set_ylim(50000,250000)
plt.legend(ncol=2, frameon=False, loc= "upper left")
plt.title("Scatter Plot: Age vs Salary by Education Level and Gender")
plt.show()
```

```python

gender_colors = {'Female':"red", "Male": "blue"}

plt.figure(figsize = (12,8))
sns.set_style("ticks")
Bachelors = sns.scatterplot(x = "Age", y = "Salary", hue = "Gender", data = Salary_data[Salary_data["Education Level"] == "Master's"], palette=gender_colors)
Bachelors.set_xlim(25,50)
Bachelors.set_ylim(50000,250000)
plt.legend(ncol=2, frameon=False, loc= "upper left")
plt.title("Scatter Plot: Age vs Salary Based on Gender Among Master's Degree Holders")
plt.show()
```

```python
gender_colors = {'Female':"red", "Male": "blue"}

plt.figure(figsize = (12,8))
sns.set_style("ticks")
Bachelors = sns.scatterplot(x = "Age", y = "Salary", hue = "Gender", data = Salary_data[Salary_data["Education Level"] == "PhD"], palette=gender_colors)
Bachelors.set_xlim(25,50)
Bachelors.set_ylim(50000,250000)
plt.legend(ncol=2, frameon=False, loc= "upper left")
plt.title("Scatter Plot: Age vs Salary Based on Gender Among PhD Degree Holders")
plt.show()
```
